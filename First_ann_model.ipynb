```python

import pandas as pd
from sklearn.preprocessing import StandardScaler # Make sure to import StandardScaler

# Assuming the file 'Churn_Modelling.csv' exists in the current directory
dataset = pd.read_csv('/Churn_Modelling.csv')

# If the file is in a different directory, provide the full path
# dataset = pd.read_csv('/path/to/your/file/Churn_Modelling.csv')

```

```python
dataset.head(3)
```

```python
dataset.isnull().sum() #check the null value in dataset
```

```python
# Print some info to check if the file was loaded correctly
dataset.info()

input_data = dataset.iloc[:,:-1]
output_data = dataset.iloc[:,-1]

ss = StandardScaler()
ss.fit_transform(input_data)
```

```python
input_data=pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
print(input_data)
```

```python
input_data
```

```python
input_data.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
import tensorflow
```

```python

from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
ann.add(Dense(units=6,input_dim=8, activation='relu'))
ann.add(Dense(units=4,activation='relu'))
ann.add(Dense(units=2,activation='relu'))
ann.add(Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train, y_train, batch_size=100, epochs=50)
```

```python
prd = ann.predict(x_test)
```

```python
prd_data = []
for i in prd:
    if i [0] > 0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
prd_data
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test, prd_data)*100
```

```python
prd1 = ann.predict(x_train)
prd_data1 = []
for i in prd1:
    if i [0] > 0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```
